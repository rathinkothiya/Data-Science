```python
import numpy as np
import pandas as pd
from plotly import __version__

import cufflinks as cf
# for using it offline.
#plotly can also be used to host online data visualizations
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
%matplotlib inline
```

```python
cf.go_offline()
```

```python
df=pd.DataFrame(np.random.rand(50,3),columns=['A','B','C'])
```

```python
#interactive plots
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B', mode='markers')
```

```python
df1=pd.DataFrame({'Sex':['Male','Male','Male','Female','Female','Female','Female'],'Values':[20,30,20,15,32,10,12]})
```

```python
df1.iplot(kind='bar',x='Sex',y='Values')
```

```python
df.iplot(kind='box')
```

```python
a=[i for i in range(0,10)]
b=[i for i in range(10,20)]
c=[np.random.randint(i) for i in range(20,30)]
df2=pd.DataFrame({'A':a,'B':b,'C':c})
```

```python
df2.iplot(kind='surface')
```

```python
df1.iplot(kind='hist')
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```

```python
df.scatter_matrix()
```
