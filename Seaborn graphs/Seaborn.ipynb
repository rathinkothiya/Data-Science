#                                            Seaborn Quick Guide

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
#dist plot
sns.distplot(tips['total_bill'])
```

```python
#dist plot without kde
sns.distplot(tips['total_bill'], kde=False)
```

```python
#changing bin value
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
#sns joint plot
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
#joint plot with kind = hex --hexagonal plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#joint plot with kind = reg --regression plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#pair plot does all the x vs y for all the numerical data in a dataset
sns.pairplot(data=tips)
```

```python
#pair plot with "hue"
#can also use palette="coolwarm"
sns.pairplot(data=tips,hue="sex")
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorical Data


```python

#Bar PLot
sns.barplot(x='sex',y='tip',data=tips)
```

```python
#bar plot with estimator
import numpy as np
sns.barplot(x='sex',y='tip',data=tips,estimator=np.std)
```

```python
#count plot
sns.countplot(x='sex',data=tips)
```

```python
#box plots
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#box plots with the another category
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#violin plot with the hue
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#Strip plot
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#swarm plot
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
#swarmplot with violin plot
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
flight=sns.load_dataset('flights')
flight.head()
```

```python
flight.corr()
```

```python
#heatmap
sns.heatmap(flight.corr())
```

```python

sns.heatmap(tips.corr(),cmap='coolwarm')
```

```python
temp=flight.pivot_table(index='month',columns='year',values='passengers')
```

```python
temp.corr()
```

```python
#cmap=magma, 
sns.heatmap(temp,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
#clustermaps
#watch with standard_scale try both the ways
sns.clustermap(temp,cmap='coolwarm',standard_scale=1)
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
#pairgrid
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
a=sns.PairGrid(iris)
a.map_diag(sns.distplot)
a.map_upper(sns.kdeplot)
a.map_lower(plt.scatter)
```

```python
#factegrid
a=sns.FacetGrid(data=tips, col='time', row='smoker')
a.map(sns.distplot,'total_bill')
```

```python
a=sns.FacetGrid(data=tips, col='time', row='smoker')
a.map(plt.scatter,'total_bill','tip')
```

```python
#regression plots
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#regression plots with hue and markers
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
#regression plots
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',markers=['v'])
```
